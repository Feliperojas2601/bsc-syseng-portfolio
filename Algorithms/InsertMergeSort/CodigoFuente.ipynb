Experimento #1: \\
Creamos una matriz llena de unos, según la potencia n de dos elegida el resultado de la suma interna debe ser nxn. 



```python
import numpy as np; 
from time import time 
def square_matrix_intern_sum_recursive(matrix): 
  n = len(matrix)
  sum = 0; 
  if n == 1: 
    sum = matrix[0,0]
  else: 
    sum = sum + square_matrix_intern_sum_recursive(matrix[0:int((n/2)),0:int((n/2))])
    sum = sum + square_matrix_intern_sum_recursive(matrix[0:int((n/2)),int((n/2)):n])
    sum = sum + square_matrix_intern_sum_recursive(matrix[int((n/2)):n,0:int((n/2))])
    sum = sum + square_matrix_intern_sum_recursive(matrix[int((n/2)):n,int((n/2)):n])
  return sum 
#Modificamos el n con potencias de 2. 
n = 128
matrix = np.ones((n,n))
print(matrix)
time_i = time()
print("La suma de los valores es: ", square_matrix_intern_sum_recursive(matrix))
time_f= time()
print(time_f-time_i)
```

```python
from random import randint
from time import time
import math
import statistics
import matplotlib.pyplot as plt
from google.colab import files 

# Implementacion del insertion sort basado en el Cormen
def insertion_sort(arr):
    for j in range(1, len(arr)):
        key = arr[j]
        i = j-1
        while i >=0 and arr[i] > key :
                arr[i+1] = arr[i]
                i = i- 1
        arr[i+1] = key

#Funcion para llenar el arreglo de tamaño n con numeros aleatorios entre 1 y 5000
def arr_filling(arr):
  for i in range(0, len(arr)):
    arr[i]= randint(1,5000)

tiempos = []
medianas = []
k = 1
timer = 0
timerI = time()
numArr = []
while (timer - timerI <= 302): 
  #Establecemos el tamaño n 
  arr = [i for i in range(1,k+1)]
  numArr.append(len(arr))

  #Establecemos la cantidad de experimentos
  num_experimentos = 100

  for i in range( 1 , num_experimentos + 1):

    arr_filling(arr)
    
    #Calculamos el tiempo de ejecucion del ordenamiento
    #print("Experimento # %d , longitud de arreglo %d " % (i, len(arr)))
    initial_time = time()
    insertion_sort(arr)
    total_time = time() - initial_time
    tiempos.append(total_time) 

    #print(*arr)
  #print(len(tiempos));
  #print(*tiempos)
  #se calcula la mediana de los tiempos para la longitud actual de arreglo y se agrega al arreglo de medianas
  #print(statistics.median(tiempos))
  medianas.append(statistics.median(tiempos))
  tiempos = []
  k += 1
  timer = time()  
    
#arreglo con medianas 
print('Medianas para insertion-sort:')
print(*medianas)
#n obtenido con un tiempo de ejecucion de 5 minutos
print(len(medianas))
#print(*numArr)

#Se hace la grafica de medianas / longitud de arreglo 
plt.plot(numArr, medianas)
plt.xlabel('n')
plt.ylabel('mediana para 100 experimentos')
plt.savefig("insertion.pdf")
files.download("insertion.pdf")
#plt.show()

#se guardan los arreglos de n y medianas para posteriormente realizar la grafica de insertion vs merge 
nInsertion = numArr
insertion = medianas


```

```python
from random import randint
from time import time
import math
import statistics
import matplotlib.pyplot as plt

# Implementacion del Merge Sort basado en el Cormen
def merge (arr,p,q,r):
  n1 = q - p + 1
  n2 = r - q
  L = [0] * (n1+1)
  R = [0] * (n2+1)
  for i in range(0 , n1 ):
    L[i] = arr[p + i]
  for j in range(0 , n2):
        R[j] = arr[q + j + 1]
  
  L[n1] = math.inf
  R[n2] = math.inf
  i = 0     
  j = 0

  for k in range(p , r+1):
        if L[i] <= R[j]:
            arr[k] = L[i]
            i = i + 1
        else:
            arr[k] = R[j]
            j = j + 1

def merge_sort (arr,p,r):
  if p < r:
    q = (p+(r-1))//2
    merge_sort(arr, p, q)
    merge_sort(arr, q+1, r)
    merge(arr, p, q, r)

#Funcion para llenar el arreglo de tamaño n con numeros aleatorios entre 1 y 5000
def arr_filling(arr):
  for i in range(0, len(arr)):
    arr[i]= randint(1,5000)

tiempos = []
medianas = []
k = 1
timer = 0
timerI = time()
numArr = []
#hasta que el tiempo de ejecucion sea mayor a 5 minutos
while (timer - timerI <= 301): 
  #Establecemos el tamaño n 
  arr = [i for i in range(1,k+1)]
  numArr.append(len(arr))
  #Establecemos la cantidad de experimentos
  num_experimentos = 100

  for i in range( 1 , num_experimentos + 1):

    arr_filling(arr)
    aux = len(arr)

    #Calculamos el tiempo de ejecucion del ordenamiento
    #print("Experimento # %d , longitud de arreglo %d " % (i, aux))
    initial_time = time()
    merge_sort (arr,0, aux-1)
    total_time = time() - initial_time 
    #el arreglo tiempos lleva los tiempos de ejecucion de los 100 experimentos para la longitud actual del arreglo
    tiempos.append(total_time)
    #print("El ordenamiento tomo un tiempo de : %.10f segundos." % total_time)

  #print(len(tiempos));
  #print(*tiempos)
  #al arreglo que lleva las medianas se agrega la mediana de los tiempos de 100 experimentos para la longitud actual del arreglo
  medianas.append(statistics.median(tiempos))
  tiempos = []
  k += 1
  timer = time()

#arreglo con medianas 
print('Medianas para merge-sort:')
print(*medianas)
#n obtenido con tiempo de ejecucion de 5 minutos 
print(len(medianas))

#Se hace la grafica de la medianas vs long arreglo para 100 experimentos 
plt.plot(numArr, medianas)
plt.xlabel('n')
plt.ylabel('mediana para 100 experimentos')
plt.savefig("merge.pdf")
files.download("merge.pdf")
#plt.show()

#se guardan los arreglos de n y medianas para posteriormente realizar la grafica de insertion vs merge 
nMerge = numArr
merge = medianas
```
