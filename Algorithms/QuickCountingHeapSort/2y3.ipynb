```python
from random import randint
from time import time
import math
import statistics
import matplotlib.pyplot as plt
from google.colab import files 

# Implementacion del Heap sort basado en el Cormen

# Heapify: Subrutina que mantiene el orden
def heapify(arr,i,heapsize):
  izq = 2*i
  der = (2*i)+1
  if izq <= heapsize-1 and arr[izq] > arr[i]:
    posmax = izq 
  else: 
    posmax = i 
  if der <= heapsize-1 and arr[der] > arr[posmax]:
    posmax = der 
  if posmax != i:
    aux = arr[posmax]
    arr[posmax] = arr[i]
    arr[i] = aux
    heapify(arr, posmax,heapsize)

# BuildHeap: Construye un Heap en un arreglo
def buildHeap(arr): 
  heapsize = len(arr)
  for i in range(int((len(arr))/2) ,0,-1):
    heapify(arr,i,heapsize)
  return heapsize
# HeapSort : Ordena el arreglo 
def heapSort(arr): 
  heapsize = buildHeap(arr)
  for i in range(len(arr)-1,1,-1): 
    aux = arr[i]
    arr[i] = arr[1]
    arr[1] = aux 
    heapsize = heapsize -1
    heapify(arr,1,heapsize)

#Funcion para llenar el arreglo de tamaño n con numeros aleatorios entre 1 y 5000
def arr_filling(arr):
  for i in range(0, len(arr)):
    arr[i]= randint(1,5000)
tiempos = []
medianas = []
k = 1
timer = 0
timerI = time()
numArr = []
while (timer - timerI <= 302): 
  #Establecemos el tamaño n 
  arr = [i for i in range(1,k+1)]
  numArr.append(len(arr))

  #Establecemos la cantidad de experimentos
  num_experimentos = 100

  for i in range( 1 , num_experimentos + 1):

    arr_filling(arr)
    
    #Calculamos el tiempo de ejecucion del ordenamiento
    #print("Experimento # %d , longitud de arreglo %d " % (i, len(arr)))
    initial_time = time()
    heapSort(arr)
    total_time = time() - initial_time
    tiempos.append(total_time) 

    #print(*arr)
  #print(len(tiempos));
  #print(*tiempos)
  #se calcula la mediana de los tiempos para la longitud actual de arreglo y se agrega al arreglo de medianas
  #print(statistics.median(tiempos))
  medianas.append(statistics.median(tiempos))
  tiempos = []
  k += 1
  timer = time()  
    
#arreglo con medianas 
print('Medianas para Heap-sort:')
print(*medianas)
#n obtenido con un tiempo de ejecucion de 5 minutos
print(len(medianas))
print(*numArr)

#Se hace la grafica de medianas / longitud de arreglo 
plt.plot(numArr, medianas)
plt.xlabel('n')
plt.ylabel('mediana para 100 experimentos')
plt.savefig("heap.pdf")
files.download("heap.pdf")
#plt.show()

#se guardan los arreglos de n y medianas para posteriormente realizar la grafica de heap vs quick 
nHeap = numArr
msHeap = medianas
```

>e. Implementación del Quick-Sort

```python
from random import randint
from time import time
import math
import statistics
import matplotlib.pyplot as plt
from google.colab import files 

# Implementacion del quick sort basado en el Cormen

# Particion: Determina una particion del arreglo utilizando un pivote. 
def particion(arr,p,r): 
  x = arr[r]
  i = p-1 
  for j in range(p,r,1): 
    if arr[j] <= x: 
      i = i+1 
      aux = arr[i]
      arr[i] = arr[j]
      arr[j]=aux
  aux = arr[i+1]
  arr[i+1] = arr[r]
  arr[r]=aux
  return i+1

#QuickSort: Utilizamos la subrutina particion para aplicar recursivamente el ordenamiento. 
def quicksort(arr,p,r): 
  if p<r: 
    q = particion(arr,p,r)
    quicksort(arr,p,q-1)
    quicksort(arr,q+1,r)

#Funcion para llenar el arreglo de tamaño n con numeros aleatorios entre 1 y 5000
def arr_filling(arr):
  for i in range(0, len(arr)):
    arr[i]= randint(1,5000)

tiempos = []
medianas = []
k = 1
timer = 0
timerI = time()
numArr = []
while (timer - timerI <= 302): 
  #Establecemos el tamaño n 
  arr = [i for i in range(1,k+1)]
  numArr.append(len(arr))

  #Establecemos la cantidad de experimentos
  num_experimentos = 100

  for i in range( 1 , num_experimentos + 1):

    arr_filling(arr)
    
    #Calculamos el tiempo de ejecucion del ordenamiento
    #print("Experimento # %d , longitud de arreglo %d " % (i, len(arr)))
    initial_time = time()
    quicksort(arr,0,len(arr)-1)
    total_time = time() - initial_time
    tiempos.append(total_time) 

    #print(*arr)
  #print(len(tiempos));
  #print(*tiempos)
  #se calcula la mediana de los tiempos para la longitud actual de arreglo y se agrega al arreglo de medianas
  #print(statistics.median(tiempos))
  medianas.append(statistics.median(tiempos))
  tiempos = []
  k += 1
  timer = time()  
    
#arreglo con medianas 
print('Medianas para quick-sort:')
print(*medianas)
#n obtenido con un tiempo de ejecucion de 5 minutos
print(len(medianas))
print(*numArr)

#Se hace la grafica de medianas / longitud de arreglo 
plt.plot(numArr, medianas)
plt.xlabel('n')
plt.ylabel('mediana para 100 experimentos')
plt.savefig("quick.pdf")
files.download("quick.pdf")
#plt.show()

#se guardan los arreglos de n y medianas para posteriormente realizar la grafica de heap vs quick 
nQuick = numArr
msQuick = medianas
```

```python
from random import randint
from time import time
import math
import statistics
import matplotlib.pyplot as plt
from google.colab import files 

# Implementacion del Counting sort basado en el Cormen

#Counting Sort: Utilizando arreglos auxiliares, impkemntamos ordenamiento en tiempo lineal.
def countingSort (arr,arrB,k): 
  arrC = [0 for i in range(k)]
  for j in range(1,len(arr)): 
    arrC[arr[j]] = arrC[arr[j]] + 1
  for i in range(1,k): 
    arrC[i] = arrC[i] + arrC[i-1]
  for j in range(len(arr)-1,0,-1): 
    arrB[arrC[arr[j]]] = arr[j]
    arrC[arr[j]] = arrC[arr[j]]-1

#Funcion para llenar el arreglo de tamaño n con numeros aleatorios entre 1 y 5000
def arr_filling(arr):
  for i in range(0, len(arr)):
    arr[i]= randint(1,100)

tiempos = []
medianas = []
k = 100
n = 100
timer = 0
timerI = time()
numArr = []
while (timer - timerI <= 302): 
  #Establecemos el tamaño n 
  arr = [0 for i in range(n)]
  numArr.append(len(arr))
  #Establecemos la cantidad de experimentos
  num_experimentos = 100

  for i in range( 1 , num_experimentos + 1):
    
    arr_filling(arr)
    arrB = [0 for i in range(len(arr))]
    
    #Calculamos el tiempo de ejecucion del ordenamiento
    #print("Experimento # %d , longitud de arreglo %d " % (i, len(arr)))
    initial_time = time()
    countingSort(arr, arrB, 101)
    total_time = time() - initial_time
    tiempos.append(total_time) 

    #print(*arr)
  #print(len(tiempos));
  #print(*tiempos)
  #se calcula la mediana de los tiempos para la longitud actual de arreglo y se agrega al arreglo de medianas
  #print(statistics.median(tiempos))
  medianas.append(statistics.median(tiempos))
  tiempos = []
  n += 1
  timer = time()  
    
#arreglo con medianas 
print('Medianas para counting-sort:')
print(*medianas)
#n obtenido con un tiempo de ejecucion de 5 minutos
print(len(medianas))
print(*numArr)
print(len(numArr))

#Se hace la grafica de medianas / longitud de arreglo 
plt.plot(numArr, medianas)
plt.xlabel('n')
plt.ylabel('mediana para 100 experimentos')
plt.savefig("counting.pdf")
files.download("counting.pdf")
#plt.show()

nCounting = numArr
msCounting = medianas


#k = 5000
#arr = [0 for i in range(k)]
#print(len(arr))
#arr_filling(arr)
#print(*arr)
#arrB = [0 for i in range(len(arr))]
#initial_time = time()
#countingSort(arr,arrB,5001)
#total_time = time() - initial_time
#print(total_time)
#print(arrB)

```

```python
# Implementacion del insertion sort basado en el Cormen
def insertion_sort(arr):
    for j in range(1, len(arr)):
        key = arr[j]
        i = j-1
        while i >=0 and arr[i] > key :
                arr[i+1] = arr[i]
                i = i- 1
        arr[i+1] = key

#Funcion para llenar el arreglo de tamaño n con numeros aleatorios entre 1 y 5000
def arr_filling(arr):
  for i in range(0, len(arr)):
    arr[i]= randint(1,5000)

tiempos = []
medianas = []
k = 1
timer = 0
timerI = time()
numArr = []
while (timer - timerI <= 302): 
  #Establecemos el tamaño n 
  arr = [i for i in range(1,k+1)]
  numArr.append(len(arr))

  #Establecemos la cantidad de experimentos
  num_experimentos = 100

  for i in range( 1 , num_experimentos + 1):

    arr_filling(arr)
    
    #Calculamos el tiempo de ejecucion del ordenamiento
    #print("Experimento # %d , longitud de arreglo %d " % (i, len(arr)))
    initial_time = time()
    insertion_sort(arr)
    total_time = time() - initial_time
    tiempos.append(total_time) 

    #print(*arr)
  #print(len(tiempos));
  #print(*tiempos)
  #se calcula la mediana de los tiempos para la longitud actual de arreglo y se agrega al arreglo de medianas
  #print(statistics.median(tiempos))
  medianas.append(statistics.median(tiempos))
  tiempos = []
  k += 1
  timer = time()  
    
#arreglo con medianas 
print('Medianas para insertion-sort:')
print(*medianas)
#n obtenido con un tiempo de ejecucion de 5 minutos
print(len(medianas))
print(*numArr)

#Se hace la grafica de medianas / longitud de arreglo 
plt.plot(numArr, medianas)
plt.xlabel('n')
plt.ylabel('mediana para 100 experimentos')
plt.savefig("insertion.pdf")
files.download("insertion.pdf")
#plt.show()

#se guardan los arreglos de n y medianas para posteriormente realizar la grafica de insertion vs merge 
nInsertion = numArr
insertion = medianas


```

```python
# Implementacion del Merge Sort basado en el Cormen
def merge (arr,p,q,r):
  n1 = q - p + 1
  n2 = r - q
  L = [0] * (n1+1)
  R = [0] * (n2+1)
  for i in range(0 , n1 ):
    L[i] = arr[p + i]
  for j in range(0 , n2):
        R[j] = arr[q + j + 1]
  
  L[n1] = math.inf
  R[n2] = math.inf
  i = 0     
  j = 0

  for k in range(p , r+1):
        if L[i] <= R[j]:
            arr[k] = L[i]
            i = i + 1
        else:
            arr[k] = R[j]
            j = j + 1

def merge_sort (arr,p,r):
  if p < r:
    q = (p+(r-1))//2
    merge_sort(arr, p, q)
    merge_sort(arr, q+1, r)
    merge(arr, p, q, r)

#Funcion para llenar el arreglo de tamaño n con numeros aleatorios entre 1 y 5000
def arr_filling(arr):
  for i in range(0, len(arr)):
    arr[i]= randint(1,5000)

tiempos = []
medianas = []
k = 1
timer = 0
timerI = time()
numArr = []
#hasta que el tiempo de ejecucion sea mayor a 5 minutos
while (timer - timerI <= 302): 
  #Establecemos el tamaño n 
  arr = [i for i in range(1,k+1)]
  numArr.append(len(arr))
  #Establecemos la cantidad de experimentos
  num_experimentos = 100

  for i in range( 1 , num_experimentos + 1):

    arr_filling(arr)
    aux = len(arr)

    #Calculamos el tiempo de ejecucion del ordenamiento
    #print("Experimento # %d , longitud de arreglo %d " % (i, aux))
    initial_time = time()
    merge_sort (arr,0, aux-1)
    total_time = time() - initial_time 
    #el arreglo tiempos lleva los tiempos de ejecucion de los 100 experimentos para la longitud actual del arreglo
    tiempos.append(total_time)
    #print("El ordenamiento tomo un tiempo de : %.10f segundos." % total_time)

  #print(len(tiempos));
  #print(*tiempos)
  #al arreglo que lleva las medianas se agrega la mediana de los tiempos de 100 experimentos para la longitud actual del arreglo
  medianas.append(statistics.median(tiempos))
  tiempos = []
  k += 1
  timer = time()

#arreglo con medianas 
print('Medianas para merge-sort:')
print(*medianas)
#n obtenido con tiempo de ejecucion de 5 minutos 
print(len(medianas))
print(*numArr)

#Se hace la grafica de la medianas vs long arreglo para 100 experimentos 
plt.plot(numArr, medianas)
plt.xlabel('n')
plt.ylabel('mediana para 100 experimentos')
plt.savefig("merge.pdf")
files.download("merge.pdf")
#plt.show()

#se guardan los arreglos de n y medianas para posteriormente realizar la grafica de insertion vs merge 
nMerge = numArr
merge = medianas
```
