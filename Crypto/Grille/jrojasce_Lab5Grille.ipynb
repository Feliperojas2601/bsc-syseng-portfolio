```python
import math
import numpy as np
import re

keySymbol = '*'

def makeKeyMatrix(key, dim2): 

  #Establish matrix and variables 
  keyMatrix = np.empty([dim2], dtype = np.dtype('U100'))

  for i in range(0, dim2): 
    if str(i) in key: 
      keyMatrix.put([i], [keySymbol]) 
    
  return keyMatrix

def rotateMatrix(keyMatrix, rotMode): 

  dim2 = len(keyMatrix)
  auxMatrix = keyMatrix.reshape(int(math.sqrt(len(keyMatrix))), int(math.sqrt(len(keyMatrix))))

  if rotMode == 1: 

    auxMatrix = np.rot90(auxMatrix)

  else: 

    auxMatrix = np.rot90(auxMatrix, 3)

  keyMatrix = auxMatrix.reshape(dim2, 1)

  return keyMatrix 

def encrypt(keyMatrix, msg, rotMode, dim2): 

  encryptedText = ''
  msgMatrix = np.empty([dim2], dtype = np.dtype('U100'))

  #Make msg matrix same as dim2
  while len(msg) < dim2 : 
    msg = msg + 'X'
 
  #Establish init, k and end to go throught the msh

  listIndex = np.where(keyMatrix == '*')
  init = 0
  k = len(listIndex[0])
  end = len(listIndex[0])

  #Evaluate indexes where is key and put values os msg 

  for i in range(0,4):
    listIndex = np.where(keyMatrix == '*')
    msgMatrix[listIndex[0]] = [letter for letter in msg[init:end]]
    keyMatrix = rotateMatrix(keyMatrix, rotMode)
    init += k
    end += k

  for letterEnc in msgMatrix: 
    encryptedText += letterEnc

  return encryptedText

def decrypt(keyMatrix, msg, rotMode, dim2): 

  decryptedText = ''

  msgMatrix = []

  msgAux = np.array([letter for letter in msg])

  #Make msg matrix same as dim2
  while len(msg) < dim2 : 
    msg = msg + 'X'

  #Establish init, k and end to go throught the msh

  listIndex = np.where(keyMatrix == '*')
  init = 0
  k = len(listIndex[0])
  end = len(listIndex[0])

  #Evaluate indexes where is key and put values os msg 

  for i in range(0,4):
    listIndex = np.where(keyMatrix == '*')
    msgMatrix.append([letter for letter in msgAux[listIndex[0]]])
    keyMatrix = rotateMatrix(keyMatrix, rotMode)
    init += k
    end += k

  for listLetterEnc in msgMatrix: 
    for letterEnc in listLetterEnc:
      decryptedText += letterEnc
    
  return decryptedText

def grille():

  #Receive dimensions of matrix 
  dim = input('Please, enter the matrix dimension: ')
  dim = dim.strip()

  #Check dim pattern as number
  pattern = re.compile('[0-9]*')
  if pattern.fullmatch(dim) == None: 
      return 'Dimension must be a valid number'

  #Calculate dimensions and manke dim integer
  dim = int(dim)
  dim2 = dim ** 2

  #Receive input, omit spaces, upper it, split by ; and validate
  key = input('Please, enter your key values (the key numbers must be separated with ;): ')
  key = key.strip()
  key = key.replace(' ','')
  keyArray = key.split(";")
  
  #Check key pattern as numbers
  pattern = re.compile('[0-9]*')
  for number in keyArray: 
    if pattern.fullmatch(number) == None or int(number) >= dim2: 
      return 'Key values must be valid numbers and not superiors to ' + str(dim2)

  #Receive message, omit spaces and upper it 
  msg = input('Please, enter your message: ')
  msg = msg.strip()
  msg = msg.replace(' ','')
  msg = msg.upper()

  #Check msg pattern
  pattern = re.compile('[A-Z]*')
  if pattern.fullmatch(msg) == None or len(msg) > dim2 : 
    return 'Message must be valid without numbers, special characters and not superior to ' + str(dim2)

  #Make msg letter to numbers
  keyMatrix = makeKeyMatrix(keyArray, dim2)

  #Receive rotation mode 1 right, 0 left
  rotMode = input('Please, enter the rotation mode (1 left, 0 right): ')
  rotMode = int(rotMode)
  if rotMode != 1 and rotMode != 0: 
    return('Invalid rotation mode.') 

  #Receive mode 1 to encrypt, 0 to decrypt
  mode = input('Please, enter the program mode (1 to Encrypt, 0 to Decrypt): ')
  mode = int(mode)
  if mode != 1 and mode != 0: 
    return('Invalid mode.') 

  #Encrypt/Decrypt and return
  if mode == 1:
    return print('Your message encrypted: ', encrypt(keyMatrix, msg, rotMode, dim2))
  else: 
    return print('Your message decrypted: ', decrypt(keyMatrix, msg, rotMode, dim2))
  
grille()
```
