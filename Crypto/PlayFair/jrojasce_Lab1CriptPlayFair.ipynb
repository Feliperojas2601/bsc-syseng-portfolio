```python
import numpy as np
import re

def makeKeyMatrix(key): 

  #Establish matrix and variables 
  keyMatrix = np.empty([5,5], dtype = str)
  letters = []
  row = 0
  col = 0
  chr_letter = ''

  #Add letters in key following order and if not repeated
  for letter in key:

    #Make letter I same to J 
    if letter == 'I' or letter == 'J': 
      letter = 'I'
 
    if letter not in letters: 
      letters.append(letter)
      keyMatrix[row][col] = letter
      if col == 4: 
        row = row + 1 
        col = 0
      else: 
        col = col + 1
  
  #Fill keyMatrix with extra letters A = 65, Z = 90
  for letter in range(65,91):

    #Convert letter to number
    chr_letter = chr(letter)

    #Make letter I same to J 
    if chr_letter == 'I' or chr_letter == 'J': 
      chr_letter = 'I'

    if chr_letter not in letters:
      letters.append(chr_letter)
      keyMatrix[row][col] = chr_letter
      if col == 4: 
        row = row + 1 
        col = 0
      else: 
        col = col + 1

  return keyMatrix 

def separateMsg(msg): 

  #Replace message letter J with I 
  msg = msg.replace('J','I')

  #Set X's of message
  i = 0
  while i < len(msg): 
    letter = msg[i]
    if i == len(msg) - 1:
      i += 2
      continue
    next_letter = msg[i+1] 
    if letter == next_letter: 
      msg = msg[:i+1] + 'X' + msg[i+1:]
    i += 2
    

  return msg

def encrypt(keyMatrix, msg): 

  encryptedText = ''

  #Make msg even 
  if len(msg) % 2 != 0: 
    msg = msg + 'X'

  # Make tuples of letters from (0,1), get indexes of letter with numpy and apply rules 
  for (letter1, letter2) in zip(msg[0::2], msg[1::2]):
        rowLetter1 = np.where(keyMatrix == letter1)[0][0]
        colLetter1 = np.where(keyMatrix == letter1)[1][0]
        rowLetter2 = np.where(keyMatrix == letter2)[0][0]
        colLetter2 = np.where(keyMatrix == letter2)[1][0]
        if rowLetter1 == rowLetter2: 
            encryptedText += keyMatrix[rowLetter1][(colLetter1 + 1) % 5] + keyMatrix[rowLetter2][(colLetter2 + 1) % 5]
        elif colLetter1 == colLetter2:
            encryptedText += keyMatrix[(rowLetter1 + 1) % 5][colLetter1] + keyMatrix[(rowLetter2 + 1) % 5][colLetter2]
        else: 
            encryptedText += keyMatrix[rowLetter1][colLetter2] + keyMatrix[rowLetter2][colLetter1]
    
  return encryptedText

def decrypt(keyMatrix, msg): 

  decryptedText = ''

  #Make msg even 
  if len(msg) % 2 != 0: 
    msg = msg + 'X'

  # Make tuples of letters from (0,1), get indexes of letter with numpy and apply rules 
  for (letter1, letter2) in zip(msg[0::2], msg[1::2]):
        rowLetter1 = np.where(keyMatrix == letter1)[0][0]
        colLetter1 = np.where(keyMatrix == letter1)[1][0]
        rowLetter2 = np.where(keyMatrix == letter2)[0][0]
        colLetter2 = np.where(keyMatrix == letter2)[1][0]
        if rowLetter1 == rowLetter2: 
            decryptedText += keyMatrix[rowLetter1][(colLetter1 - 1) % 5] + keyMatrix[rowLetter2][(colLetter2 - 1) % 5]
        elif colLetter1 == colLetter2:
            decryptedText += keyMatrix[(rowLetter1 - 1) % 5][colLetter1] + keyMatrix[(rowLetter2 - 1) % 5][colLetter2]
        else: 
            decryptedText += keyMatrix[rowLetter1][colLetter2] + keyMatrix[rowLetter2][colLetter1]
    
  return decryptedText

def playFair():

  #Receive input, omit spaces, upper it and validate
  key = input('Please, enter your key: ')
  key = key.strip()
  key = key.replace(' ','')
  key = key.upper()
  if len(key) > 25: 
    return 'Key must contain 25 characters or less.'
  
  #Check key pattern
  pattern = re.compile('[A-Z]*')
  if pattern.fullmatch(key) == None: 
    return 'Key must be valid without numbers and special characters'

  #Generate Key matrix  
  keyMatrix = makeKeyMatrix(key)

  #Receive message, omit spaces and upper it 
  msg = input('Please, enter your message: ')
  msg = msg.strip()
  msg = msg.replace(' ','')
  msg = msg.upper()

  #Check msg pattern
  if pattern.fullmatch(msg) == None: 
    return 'Message must be valid without numbers and special characters'


  #Receive mode 1 to encrypt, 0 to decrypt
  mode = input('Please, enter the program mode (1 to Encrypt, 0 to Decrypt): ')
  mode = int(mode)
  if mode != 1 and mode != 0: 
    return('Invalid mode.') 

  #Separate message letters if same letter concatenated 
  if mode == 1:
    msg = separateMsg(msg)

  #Encrypt/Decrypt, replace I for I/J and return
  if mode == 1:
    return print('Your message encrypted: ', encrypt(keyMatrix, msg).replace('I','I/J'))
  else: 
    return print('Your message decrypted: ', decrypt(keyMatrix, msg).replace('I','I/J'))
  
playFair()
```
